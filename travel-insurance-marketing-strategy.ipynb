***Travel insurance customer analysis:**

A company named 'ABC' is into travel insurance business. Due to coid-19, their business got affected.
So, Company wants to reduce its marketing expenditure by 50%.

Company has previous data about customers. The data set contains columns like,

'Age' - represents customer age
'Employment Type' - represents customer employement sector (private/government)
'GraduateOrNot' - represents whether customer graduated or not
'AnnualIncome' - shows customer annual income in dollars
'FamilyMembers' - shows number of depedents family members of customer
'ChronicDiseases' - shows whether customer have chroni disease or not (1 is suffering with diesease)
'FrequentFlyer' - shows whether customer is a frequent flyer or not
'EverTravelledAbroad' - shows whether customer travelled abroad or not
'TravelInsurance' - shows whether customer has taken insurance or not (1 is insurance taken)

company wants to reduce marketing expenditure by advertising to correct group of people.*

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset into pandas dataframe

df=pd.read_csv('../input/travel-insuran/travel_insurance.csv')
df.info()
```

> Dataset contains 1987 records and 8 features without any missing values.

```python
# Statistical overview of the data
print('Statistical overview of numerical features: \n\n',df.describe())
print('\n\nStatistical overview of categorical features: \n\n',df.describe(exclude='number'))
```

> Stastical overview indicates that TravelInsurance & ChronicDiseases features also of categorical type.

> For analysis, we could convert remaining features like age, AnnualIncome & FamilyMembers into categorical type.
> Age grouped into 2 categories. (25-30 years & above 30 years)
> AnnualIncome grouped into 2 categories. (300000-900000 & above 900000)
> FamilyMenbers grouped into 2 categories. (2-5 & above 5)

```python
# converting numerical features into different categories

for i in df.index:
    if df.loc[i,'Age']<=30:
        df.loc[i,'Age']='25-30'
    else:
        df.loc[i,'Age']='Above 30'

for i in df.index:
    if df.loc[i,'AnnualIncome']<=900000:
        df.loc[i,'AnnualIncome']='300000-900000'
    else:
        df.loc[i,'AnnualIncome']='Above 900000'

for i in df.index:
    if df.loc[i,'FamilyMembers']<=5:
        df.loc[i,'FamilyMembers']='3-5'
    else:
        df.loc[i,'FamilyMembers']='Above 5'
```

> Now will analyse each categorical variable to fins out its impact on travel insurance business
> we will consider TravelInsurance being 1 as the customer and 0 as the non_customer of the company

```python
def count_pct(data,col_name,measure,category):
    
    '''
    This function will results in a bar plot of % of people as customer & non_customer for 
    each category of mentioned categorical feature
    
    data : dataframe of interest
    col_name : Categorical column of data (interested categorical feature)
    measure : measure/output column of dataframe (in this case TravelInsurance column)
    category : string to represnts the title of the chart
    
    '''
    
    df_emp1=data.groupby([col_name]).size().to_frame(name='count').reset_index()
    df_emp=data.groupby([col_name,measure]).size().to_frame(name='count').reset_index()
    df_emp[measure]=(df_emp[measure].map({0:'non_customer',1:'customer'})).astype('str')
  

    for i in df_emp.index:
        if df_emp.loc[i,col_name]=='No' or df_emp.loc[i,col_name]==0:
            df_emp.loc[i,'category']= 'No '+category
        elif df_emp.loc[i,col_name]==1 or df_emp.loc[i,col_name]=='Yes':
            df_emp.loc[i,'category']= category
        else:
            df_emp.loc[i,'category']= df_emp.loc[i,col_name]
        for j in df_emp1.index:
            if df_emp.loc[i,col_name]== df_emp1.loc[j,col_name]:
                df_emp.loc[i,'%']=(df_emp.loc[i,'count']*100/df_emp1.loc[j,'count']).round(1)
                df_emp.loc[i,'category']=df_emp.loc[i,measure]+' & \n'+ df_emp.loc[i,'category']
        
        sns.set_style('darkgrid')
        sns.set_palette('deep')
        sns.set(font_scale=1.4)
        pic=sns.barplot(data=df_emp,x='category',y='%')
        pic.set_xticklabels(labels=pic.get_xticklabels(),rotation=60,ha='right')
        pic.set_ylabel('% of people')
        pic.set_xlabel('')
        pic.set_title(category+' vs Insurance',weight='bold')
        pic.figure.set_size_inches(12,30)
        plt.tight_layout()           
    return pic
    
```

```python
fig=plt.figure()

plt.subplot(421)
age_insurance = count_pct(df,'Age','TravelInsurance','Age')
plt.subplot(422)
emp_type_insurance = count_pct(df,'Employment Type','TravelInsurance','Employment Type')
plt.subplot(423)
graduate_insurance = count_pct(df,'GraduateOrNot','TravelInsurance','Graduate')
plt.subplot(424)
income_insurance = count_pct(df,'AnnualIncome','TravelInsurance','Annual Income')
plt.subplot(425)
family_memb_insurance = count_pct(df,'FamilyMembers','TravelInsurance','Family Members')
plt.subplot(426)
chronic_disease_insurance = count_pct(df,'ChronicDiseases','TravelInsurance','Chronic Diseases')
plt.subplot(427)
freq_flyer_insurance = count_pct(df,'FrequentFlyer','TravelInsurance','Frequent Flyer')
plt.subplot(428)
abroad_travel_insurance = count_pct(df,'EverTravelledAbroad','TravelInsurance','Abroad Travel')
plt.subplots_adjust(hspace=1.6)
```

> People with abroad travel history or frequent flying history or annual income above 900000 dollars 
are having a probability of minimum 50% to be company's customers
> all other categories are having probability less than 50%

> now we will analyse effect of two features together on travel insurance business

> Now we will analyze effect all three important features together on travel insurance business

```python
# abroad travel & annual income & frequent travel vs travel insurance

ax=sns.catplot(data=df,y='TravelInsurance',x='AnnualIncome',hue='EverTravelledAbroad',
               col='FrequentFlyer',kind='bar',legend_out=True)
ax.axes[0,0].set_xlabel('Annual Income')
ax.axes[0,1].set_xlabel('Annual Income')
ax.axes[0,0].set_ylabel('% of people insured')
```

1. With high confidense we could say that people with annual income above 900000 dollars & abroad  travel history & frequent travel history are having an average probability of 90% to be company's customers

2.  With high confidense we could say that people with annual income above 900000 dollars & abroad  travel history & no frequent travel history are having an average probability of 80% to be company's customers

```python
# extracting records of people with abroad travel or annual income above 900000 dollars 
# or frequent travel

df1=df[(df['AnnualIncome']=='Above 900000') | (df['EverTravelledAbroad']=='Yes')
       | (df['FrequentFlyer']=='Yes')]
df_targeted=df1.groupby('TravelInsurance').size().to_frame(name='count').reset_index()
df_targeted['TravelInsurance']=df_targeted['TravelInsurance'].map({0:'Non-Customer',1:'Customer'})

x=['Overall group','Targeted group']
y=[df.shape[0],df1.shape[0]]
plt.bar(x,height=y)
plt.title('Number people to be advertised')
```

1. As shown above, targeted group people number is almost reduced to half. As a result company's marketing advertisement cost reduces

```python
fig,ax=plt.subplots(1,2,figsize=(12,10))


ax[0].pie(x='Age',data=df.groupby('TravelInsurance').count(),
        labels=df.groupby('TravelInsurance').count().index.map({0:'Non-Customer',1:'Customer'}),
        autopct='%0.2f%%')
ax[0].title.set_text('Overall Customer Distribution')


ax[1].pie(x='count',data=df_targeted,
        labels=df_targeted['TravelInsurance'],
        autopct='%0.2f%%')
ax[1].title.set_text('Targeted Customer Distribution')
```
